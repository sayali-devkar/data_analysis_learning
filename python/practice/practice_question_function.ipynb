### 3. Write a function to find the maximum of two numbers.

```python
def max_of_two_num():
    n1 = int(input("Enter the number : "))
    n2 = int(input("Enter the number : "))
    if n1 > n2 : 
        print(f"{n1} is greater than {n2}")
    else:
        print(f"{n2} is greater than {n1}")
```

### 4. Write a function to calculate the area and perimeter of a circle.

```python
def area_peri_circle():
    radius = int(input("Enter the radius : "))
    area = 3.14*radius*radius
    perimeter = 2*3.14*radius
    print(f"Area of circle is {area}")
    print(f"Perimeterr of circle is {perimeter}")
```

### 7. Write a function to find the factorial of a number.

```python
def factorial(n):
    result = 1
    for i in range(1 , n+1):
        result = result*i
    print(result)
```

```python
n = 5
factorial(n)
```
