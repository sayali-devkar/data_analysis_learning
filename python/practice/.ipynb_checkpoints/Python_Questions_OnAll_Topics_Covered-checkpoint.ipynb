### Level 1: Data Types & Conditionals
#### 1. Ask the user to input a number.
- Print whether it’s even or odd.
- Print whether it’s positive, negative, or zero.

```python
# 1 even and odd
num = int(input("Enter the number to check even or odd : "))
if num % 2==0 : 
    print(f" {num} is even number")
else : 
    print(f"{num} is odd number" )
    
```

```python
# check whether the number is psoitive/negative/zero
num = int(input("Enter the number to check even or odd : "))
if num > 0 : 
    print(f"{num}is a positive number")
elif num < 0 :
    print(f"{num} is a negative number")
elif num ==0 : 
    print("Entered number is a zero")
```

### 2. Ask the user to input a string.
- Print length of the string.
- Print the string in reverse.
- Check if the string is a palindrome (same forward and backward).

```python
name = input("Enter your name : ")
print("Length of your name is : ",len(name))
print("Reverse of your name is  : ", name[::-1])
```

```python
name = input("Enter your name : ")
if name == name[::-1]:
    print("Your name is a palindrome!")
else : 
    print("Beautifull name !! Dear your name is not a palindrome") 
```

### Level 2: Loops & Lists

### 3. Ask the user for n numbers and store them in a list.

- Print the sum of all numbers.

- Print the largest and smallest number.

- Print all even numbers from the list.

### 4. Print the multiplication table of a number using a for loop.

### 5. Print a triangle pattern using * for a given number of rows using a while loop.

```python
numbers = []
n = int(input("How many numbers do you want to insert in list : "))
for i in range(1 ,n+1):
    num = int(input("Enter the number : "))
    numbers.append(num)
print(numbers)
even_numbers = list(filter(lambda a : a%2==0 , numbers))
print(even_numbers)
maximum_from_list = list(filter(lambda a : a > n , numbers))
print("Largest number from list is :" , maximum_from_list)
minimum_from_list = list(filter(lambda a : a < n , numbers))
print("Smallest number from list is : " , minimum_from_list)
    
```

```python
#4 
for i in range (1 , 11):
    print(5*i)
```

### Level 3: Tuples, Sets, Dictionaries
### 5. Ask the user for a list of numbers.

Convert it to a tuple and print it.

Convert it to a set and print it (check duplicates).

### 6. Ask the user for 5 key-value pairs and store them in a dictionary.

Print all keys, all values, and the items.

Ask the user for a key and print its value if it exists, else print “Key not found”.

```python
numbers = []
n = int(input("Enter how many numbers you want : "))
for i in range(1 , n+1):
        num = int(input("Enter the number : "))
        numbers.append(num)
print(numbers)
t = tuple(numbers)
print(t)
s = set(numbers)
print(s)
```

```python
d ={}
number_of_key_value_pair_you_want = int(input("Enter the number of pairs you want : "))
for i in range( 1, number_of_key_value_pair_you_want +1):
    k_input = input("Enter the key : ")
    v_input = input("Enter the value : ")
    d[k_input] = v_input
print(d)
```

```python
l = []
number_of_elements_in_list = int(input("Enter the number of element in list : "))
for i in range (1 , number_of_elements_in_list):
    elements = int(input("Enter the elements : "))
    l.append(elements)
print(l)
even = list(filter(lambda a : a%2==0 , l))
print(even)
```
