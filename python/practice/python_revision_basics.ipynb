## 🔹 **Task 2: Square then filter**
- Given:
- nums = [10, 15, 20, 25, 30]
- First: square all numbers
- Then: keep only numbers greater than 400

```python
# Square of all numbers 
nums = [10,15,20,25,30]
nums = list(map(lambda a : a**2 , nums))
print(nums)
```

```python
# filtering numbers greating than 400
nums = list(filter(lambda a: a>400 , nums))
print(nums)
```

## 🔹 **Task 3: Frequency count using dictionary**

```python
items = ['a', 'b', 'a', 'c', 'b', 'a']
def frequencyc() :
    freq={}
    for i in items : 
        if i in freq:
            freq[i] += 1
        else : 
            freq[i] = 1
    return(freq)
result = frequencyc()
print(result)
```
