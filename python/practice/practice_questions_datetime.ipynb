### Q1  How do you get the current date and time in Python?

```python
from datetime import datetime 
```

```python
now = datetime.now()
now
```

```python
# Formatted date and time using strftime()
```

```python
now = datetime.now()
formatted_now = now.strftime("%d-%m-%Y %H:%M:%S")

print(formatted_now)
```
