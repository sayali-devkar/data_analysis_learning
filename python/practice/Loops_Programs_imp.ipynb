### Q1. Print numbers from 1 to 10 using while

```python
i = 1
while i  < 11 : 
    print(i)
    i = i+1
```

### Q3 Print numbers from 10 to 1 (reverse)

```python
i = 10
while i >= 1:
    print(i)
    i = i-1
```
