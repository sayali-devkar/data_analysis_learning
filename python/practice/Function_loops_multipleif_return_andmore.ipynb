```python
salaries = [26, 27, 28, 29, 30, 30, 31, 29, 28, 27, 30, 31, 32, 29, 28, 95]

def analyze_salary(salaries):
    total= 0
    sum_low = 0
    for i in salaries:
        if i <= 35:  # check current value, not list
            total += 1               # count
            sum_low += i             # sum of salaries
    return sum_low, total           # return both values

# call the function
sum_low, count_low = analyze_salary(salaries)
average = sum_low / count_low

print("Average salary below 35:", average)
```

### Write a function clean_salaries that:

- 1️⃣ Creates a new list of salaries below 35
- 2️⃣ Counts how many salaries were removed (≥35)
- 3️⃣ Returns:
- clean list
- removed count

```python
salaries = [24, 26, 28, 30, 31, 33, 29, 27, 95]
salaries_below_35 = []
def clean_salaries(salaries):
    remove = 0
    for i in salaries:
        if i < 35 :
            salaries_below_35.append(i)
        else : 
            remove = remove+1
    return(remove)
result = clean_salaries(salaries)
print("Salaries below 35 are : " , salaries_below_35)
print("Number of salaries removed:", result)
```
