### agar sare elements chahiye to use map if u want to see the result of all elements 

```python
lst1 = [1,2,3,4,-65,-78,54,-78,-908]
pos_nega = list(map(lambda a  :  a > 0 ,lst1))
print(pos_nega)
```

### double even elements using lambda

```python
lst1 = (1,2,3,4,5)
double_even = tuple(map(lambda a  : a*2 if a%2==0 else a , lst1))
print(double_even)
```

### 4 – lambda returning boolean
Write a lambda that:
takes a number
returns True if it’s divisible by both 3 and 5
else False

```python
a = int(input("Enter the number : " )) 
divisibility = lambda a : "True" if a%3==0 and a%5==0 else "False"
print(divisibility(a))
```

```python
a = int(input("Enter the number : " )) 
divisibility = lambda a :   a%3==0 and a%5==0 
print(divisibility(a))
```

```python
names = ["ram", "shyam", "mohan"]
upper_case = list(map(lambda a : a.upper() , names))
print(upper_case)
```

Create a new list where:

numbers divisible by 3 are multiplied by 10

all other numbers remain unchanged

```python
nums = [1, 2, 3, 4, 5, 6]
ss = list(map(lambda a  : a*10 if a%3==0 else a , nums))
print(ss)
```

```python
nums = [-5, 0, 7, -2, 9, -1]
num_greater_than_zero_equal_to_zero = list(filter(lambda a : a >= 0 , nums))
print(num_greater_than_zero_equal_to_zero)
```
