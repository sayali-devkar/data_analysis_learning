### timedelta → TIME DIFFERENCE ➕➖

```python
from datetime import date , timedelta
```

```python
todays_date = date.today()
print(todays_date)
date_after_3_years = todays_date + timedelta(days=3*365)  #here 3 is just a multilication factor so it give 1095 it considers it as days beacuse 
print(date_after_3_years)                                 # first parameter of timedelta is days so it considers 1095 as days that is 3 years.so we get date after 3 years 
```

```python
date_before_7_weeks = todays_date - timedelta(weeks=7)
print(date_before_7_weeks)
```

### strftime()

- used to convert the date to a string.
- we cannot perform date math on the output as it will be a string.
- ex. 16/jan/2026

##### syntax = date_or_datetime.strftime("FORMAT")

```python
from datetime import datetime
```

```python
now = datetime.now()
now
```

```python
print(now.strftime("%Y-%m-%d %H:%M:%S"))
```

```python
print(now.strftime("%Y-%m-%d"))
```

```python
print(now.strftime("%H:%M:%S"))
```

```python
print(now.strftime("%d-%B-%Y"))
```

```python
print(now.strftime("%I:%M:%S %p"))
```

```python
print(now.strftime("%A, %d-%B-%Y"))
```
