### 1. WHILE LOOP

#### while loop with continue

```python
i =0  #initialization
while i <10:
    i = i+1 #increment
    square = i**2
    if 20 < square < 40:
        print("********Iterations skipped********")
        continue
    print(f"square of {i} is {square}")
```

```python
n = int(input("Enter a number"))
i = 1 #initialiization
factors = 0  # no of facctors
while i <= n:
    if n%i == 0:
        factors = factors+1
    i = i+1 #increment
if factors == 2:
    print(f"{n} is a prime number")
else:
    print(f"{n} is a non-prime number")
    
```
