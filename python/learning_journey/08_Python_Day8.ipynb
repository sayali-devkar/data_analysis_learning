```python
country = input("Name of the country : ")

if country == country.lower():  # check krnar ki enter keleli country lower case mdhe aahe ka nahi jr nasel tr khich return nhi honar.
    print("The the font enter is all lower case")

elif country == country.upper():
    print("The font entered is all upper case")

elif country == country.title():
    print("The font is of title case")

else:
    print("The font enetered is of mixed case")
```

### Nested statements

- IF INSIDE A IF STATEMENT

```python
marks = int(input("Enter the marks :"))
if marks >= 40 :  #outer if
    if marks >= 75 : #inner if
        print("pass with distinction")
    else :    #inner else
        print("pass")
else:        #outer else
    print("Fail")
```
