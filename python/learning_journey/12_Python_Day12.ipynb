#####  Parameter vs argument

- A parameter is a variable listed inside the paranthesis in a function defination.
- A argument is a value that is sent to a function when it is called.

```python
def square(list1):   #list1 >>> parameter 
    list1 = [1,2,3,4]
    square_list = [i**2 for i in list1]
    print(square_list)
```

```python
list1 = [1,2,3,4]  # list1 is a parameter which is given at the time of 
square(list1)
```

```python
list2 = [11,21,23]
square(list2)    # list 2 is an argument bacuse its gven at run time
```

```python
def area_peri_of_rectangle(length,breadth):
    area = length*breadth
    perimeter = 2*(length+breadth)
    print("Area of rectangle is : " , area)
    print("Perimeter of rectangle is : " ,perimeter) 
```

```python
area_peri_of_rectangle(20,10)
```
