### Dictionary

- Dictionary is a mutable data type.
- Dictionary is mutable which means we can perform add,delete,update and pop methods.
- when a data type is immutable such as tuple we cannot perform such operations on it.

```python
d = {}           # This is a empty dictionary.
print(d,type(d))
```

```python
marks = {'phy':89,'maths':67 ,'chem':56}        #1 pair is consider as i whole index
print(marks,type(marks))
len(marks)
```

```python
marks[0]
```

- Access using key 

```python
marks['phy']
```

```python
# To access all keys at a time :->
marks = {'phy':89,'maths':67 ,'chem':56}  
marks.keys()
```

```python
# To access all values at a time :->
marks = {'phy':89,'maths':67 ,'chem':56}  
marks.values()
```

```python
# To access all items at a time :->
marks = {'phy':89,'maths':67 ,'chem':56}  
marks.items()
```

### FUNCTIONS OF DICT : 

#### 1. Add and Update

```python
# ADD
marks = {'phy':89,'maths':67 ,'chem':56}
marks['IT'] = 99         # ADDS ONE ELEMENT AT LAST
print(marks)
```

```python
# UPDATE
marks = {'phy':89,'maths':67 ,'chem':56}
marks.update({'eng':90})
print(marks)
```

```python
# UPDATE : USED TO ADD MULTIPLE ELEMENTS TOO
marks = {'phy':89,'maths':67 ,'chem':56}
marks.update({'eng':90,'evs':88})
print(marks)
```

```python
# UPDATING DICT ELEMENTS
marks = {'phy':89,'maths':67 ,'chem':56}
marks['phy']=76
print(marks)
```

### 2.POP

```python
marks = {'phy':89,'maths':67 ,'chem':56}
marks.pop('chem')  #remove specified key and return the corresponding value. 
print(marks)            # we have to give its not like the list removes only the last element.
```

### 3.Clear

```python
marks = {'phy':89,'maths':67 ,'chem':56}
marks.clear()
print(marks)
```

### 4. Delete

```python
marks = {'phy':89,'maths':67 ,'chem':56}
del marks
print(marks) #below we got the error which means dict deleted successfully.
```

```python
marks = {'phy':89,'maths':67 ,'chem':56}
del marks['chem']
print(marks)
```
