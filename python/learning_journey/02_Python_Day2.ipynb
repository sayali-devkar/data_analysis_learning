## TYPE CASTING

1. int to float >>>> possible

```python
a = 10
print(a,type(a)) 
b = float(a) #TYPECASTING OF A FROM INT TO FLOAT
print(b,type(b))
```

2. int to complex >>>> possible

```python
a = 10
print(a,type(a)) 
b = complex(a) #TYPECASTING OF A FROM INT TO FLOAT
print(b,type(b))
```

3. float to int >>> possible

```python
a = 10.0
print(a,type(a)) 
b = int(a) #TYPECASTING OF A FROM INT TO FLOAT
print(b,type(b))
```

4. float to complex

```python
a = 10.0
print(a,type(a)) 
b = complex(a) #TYPECASTING OF A FROM INT TO FLOAT
print(b,type(b))
```

5. complex to int  >>>> not possible data type error

```python
a = 10+0j
print(a,type(a)) 
b = int(a) #TYPECASTING OF A FROM INT TO FLOAT
print(b,type(b))
```
