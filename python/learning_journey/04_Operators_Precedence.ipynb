# FOLLOWS "PEMDAS" 

```python
# Parentheses () – Highest precedence, operations inside parentheses are evaluated first.
# Exponents ** – Power calculations (e.g., 2 ** 3 → 8).
# Multiplication *, Division /, Floor Division //, Modulus % – Evaluated from left to right.(if first is division while evaluating from left t right
# then do it then do others and vie versa for multiplication they all same priority.
# Addition +, Subtraction - – Evaluated from left to right.
```

```python
result = 10 + 2 * 3  # Multiplication happens first: 10 + (2 * 3) = 16
print(result)
```

```python
result = (10+2)*3
print(result)
```

```python
result = 2 ** 3 ** 2  # Right-to-left exponentiation: 2 ** (3 ** 2) = 2 ** 9 = 512
print(result)
```

```python
result = 10 + 6 / 3 * 4 - 2 ** 2
print(result)
```

```python
result = 18 // 4 + 3 * 2 ** 2 % 5
print(result)
```
