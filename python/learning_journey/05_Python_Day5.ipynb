#### Reverse and sort a list

```python
list1 = [10,20,30,40,50]
list2 = list1[::-1] #Negative step size (Reverse by using slicing)
print(list1)
print(list2)
```

```python
list1 = [10,20,30,40,50]
list1.reverse()
print(list1)
```

```python
list1 = [1,4,2,6,8,4,3,2,1,0]
list1.sort()      
#Ascending   (Mutable)
print(list1)
```

```python
list1 = [1,4,2,6,8,4,3,2,1,0]
list1.sort(reverse=True)      
print(list1)
list1
```

```python
list1 = [1,4,2,6,8,4,3,2,1,0]
print("Original list :", list1)
list2 = sorted(list1)   
#ascending
print("Sorted list :", list2)
```

```python
list1 =  [1,4,2,6,8,4,3,2,1,0]
print("Original list : " , list1)
list2 = sorted(list1,reverse=True) #descending
print("Sorted list :", list2)
```

```python
list1 = [1,2,3, 8.6, 5.6]
list1.sort()
print(list1)
```

```python
list1 = [1,2,3, 8.6, 5.6, 9+6j]
list1.sort()
print(list1)
```

### All and Any

```python
list1 = [1,2,3,4,5]
all(list1)
```

```python
list1 = [1,2,3,4,5,0]
all(list1)
```

```python
list1 = [1,2,3,4,5, False]
all(list1)
```

```python
list1 = [1,2,3,4,5]
any(list1)
```

```python
list1 = [1,2,3,4,5,0]
any(list1)
```

```python
list1 = [1,2,3,4,5, False]
any(list1)
```

```python
list1 = [False, 0, 1]
any(list1)
```

```python
list1 = [False, 0]
any(list1)
```
