```python
import pandas as pd
```

```python
df = pd.DataFrame({
    "Department": ["HR", "HR", "IT", "IT", "Marketing", "Marketing", "Sales", "Sales"],
    "Team": ["A", "A", "B", "B", "C", "C", "D", "D"],
    "Gender": ["M", "F", "M", "F", "M", "F", "M", "F"],
    "Salary": [85, 90, 78, 85, 92, 88, 75, 80],
    "Age": [23, 25, 30, 22, 28, 26, 21, 27],
    "JoinDate": pd.to_datetime([
        "2020-01-10", "2020-02-15", "2021-03-20", "2021-04-10",
        "2020-05-30", "2020-06-25", "2021-07-15", "2021-08-01"
    ])
})  
df
```

## This says:
### **“Group by Department, then calculate average Salary for each group.”**

```python
df.groupby("Department")["Salary"].mean()
```

```python
type(df.groupby("Department")["Salary"].mean())
```

### **HOW MUCH MONEY IS GOING INTO SALARY OF EACH DEPARTMENT**

```python
df.groupby("Department")["Salary"].sum()
```

### **HOW MANY PEOPLE ARE WORKING IN EACH TEAM**

```python
df.groupby("Team")["Salary"].count()
```

### **MAXIMUM SALARY FROM DEPARTMENT**

```python
df.groupby("Department")["Salary"].max()
```

### **MINIMUM SALARY FROM DEPARTMENT**

```python
df.groupby("Department")["Salary"].min()
```

## Custom Aggregations with .agg()
- **Apply multiple functions at once like this:**

```python
df.groupby("Team")["Salary"].agg(["mean", "max", "min"])
```

```python
type(df.groupby("Team")["Salary"].agg(["mean", "max", "min"]))
```

```python
df.groupby("Team")["Salary"].agg(
    avg_score="mean",
    high_score="max"
)
```

```python
df.groupby("Team").agg({
    "Salary": "mean",
    "Age": "max"
})
```

## **transform()** Example: **Same shape as original** . **Add new column based on group**

```python
df["Team Avg pagar"] = df.groupby("Team")["Salary"].transform("mean")
df
```

## **filter()**	Subset of rows. **Keep/discard whole groups**

```python
df.groupby("Team").filter(lambda x: x["Salary"].mean() > 85)
```
