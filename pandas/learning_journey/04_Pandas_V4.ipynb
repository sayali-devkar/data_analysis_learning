## Question:
- Filter movies that satisfy EITHER of the following:
- Condition A:
- **Genre** is **Horror Comedy**
- **IMDb rating** is **greater than 6**
- OR
- Condition B:
- **BoxOffice(INR Crore) is more than 500**

```python
df[(df["Genre"] == "Horror Comedy") & (df["IMDb"] > 6) | (df["BoxOffice(INR Crore)"] > 500)]
```

```python
# cleaner code for the same task
df[
    (
        (df["Genre"] == "Horror Comedy") &
        (df["IMDb"] > 6)
    )
    |
    (df["BoxOffice(INR Crore)"] > 500)
][["Actor", "Film", "Genre", "IMDb", "BoxOffice(INR Crore)"]]
```
