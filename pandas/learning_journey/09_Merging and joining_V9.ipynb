```python
import pandas as pd
```

```python
employees = pd.DataFrame({
    "EmpID": [1, 2, 3],
    "Name": ["Alice", "Bob", "Charlie"],
    "DeptID": [10, 20, 30]
})
```

```python
employees
```

```python
departments = pd.DataFrame({
    "DeptID": [10, 20, 40],
    "DeptName": ["HR", "Engineering", "Marketing"]
})
```

```python
departments
```

```python
pd.merge(employees , departments ,on = "DeptID")        #Inner Join
```

# *Left Join* 
- **JO BHI EMP TABLE KE ROWS HAI WO SAB AYENGE AND MATHCHING FROM DEPT TABLE** 

```python
pd.merge(employees , departments , on="DeptID",how="left")
```

## *Right Join*
- **jo right table hai uske sare records de do and jo emp me available nhi hai usko nan return kro**

```python
pd.merge(employees , departments , on="DeptID" , how = "right")
```

## outer join
- **sare dept id's mil jayenge and jo empty hoga waha pe nan aa jayega**

```python
pd.merge(employees , departments , on="DeptID" , how = "outer")
```
