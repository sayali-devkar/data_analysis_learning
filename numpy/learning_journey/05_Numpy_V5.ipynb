### **3. Why data Types matter in Numpy.**
  1. Memory Usage : smaller data types use less memeory
  2. Performance : Operations on smaller data type are faster due to less data being processed.
  3. Precision : Choose appropriate data type to ensure that we dont loose that precision(ex use float32 istead of float640
